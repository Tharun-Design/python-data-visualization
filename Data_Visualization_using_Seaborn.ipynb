Why visualize data?


*   Observe the patterns
*   Identify extreme values that could be anomalies
*   Easy interpretation

**Importing necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Scatter plot**

Scatter plot of Price vs Age with default arguments

```python
#Importing data
cars_data = pd.read_csv("Toyota.csv", index_col=0, na_values=["??","????"])

#Removing missing values from the dataframe
cars_data = cars_data.dropna(axis=0, inplace= False)
cars_data
```

```python
sns.set(style = 'darkgrid')
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], color = 'red')
```

```python
#Scatter plot of Price vs Age without the regression fit line
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], fit_reg = False)
```

```python
#Scatter plot of Price vs Age by customizing the appearance of markers
sns.regplot(x = cars_data['Age'], y = cars_data['Price'], fit_reg = False, marker = '*')
```



*   Scatter plot of Price vs Age by FuelType
*   Using hue parameter, including another variable to show the fuel
types categories with different colors






```python
sns.lmplot(x = 'Age', y = 'Price', data = cars_data, fit_reg = False, hue = 'FuelType', legend = True, palette = 'Set1')
```

**Histogram**

```python
#Histogram with default kernel density estimate
sns.distplot(cars_data['Price'])
```

```python
#Histogram without kernel density estimate
sns.distplot(cars_data['Price'], kde = False)
```

```python
#Histogram with fixed no. of bins
sns.distplot(cars_data['Price'], bins = 10)
```

**Bar Plot**

```python
#Frequency distribution of fuel type of the cars
sns.countplot(x = 'FuelType', data = cars_data)
```

```python
#Grouped bar plot of FuelType and Automatic
sns.countplot(x = 'FuelType', data = cars_data, hue = 'Automatic')
```

**Box and whiskers plot**
Box and whiskers plot of Price to visually interpret the
five
number summary

```python
sns.boxplot(x = 'Price', data = cars_data)
```



*   Box and whiskers plot for numerical vs categorical variable
*   Price of the cars for various fuel types






```python
sns.boxplot(x = 'FuelType', y = 'Price', data = cars_data)
```

Grouped box and whiskers plot of Price vs FuelType and Automatic

```python
sns.boxplot(x = 'FuelType', y = 'Price', data = cars_data, hue = 'Automatic')
```

**Pairwise plots**


*   It is used to plot pairwise relationships in a dataset
*   Creates scatterplots for joint relationships and histograms for
univariate distributions


```python
sns.pairplot(cars_data, kind = 'scatter', hue = 'FuelType')
```
